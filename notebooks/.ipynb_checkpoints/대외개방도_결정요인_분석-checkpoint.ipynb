## 1. 대외개방도

### 경제규모 대비 무역액 비율 측정

한 나라의 세계 경제 통합 정도를 가장 자연스럽게 측정할 수 있는 지표는 개방도(openness)이다. 어떤 나라 $i$의 총수출($X_i$), 총수입($M_i$), GDP($Y_i$)를 사용해 $i$국의 **대외개방도**를 계산하면 다음과 같다.

$$
O_i = \frac{X_i + M_i}{Y_i}
$$

이를 **대외의존도**라 부르기도 한다. $O_i$ 값이 높을수록 해당 국가는 더 개방적이다. 싱가포르와 같은 작은 개방 경제의 경우, 이 개방도 지표는 1을 훨씬 초과할 수도 있다. 이 개방도 지표는 시간에 따라 추적할 수 있다. 예를 들어, [Penn World Tables(PWT)](https://www.rug.nl/ggdc/productivity/pwt/)는 여러 연도를 포괄하는 개방도 지표를 제공한다.

그러나 $O_i$를 국가 간 비교에 그대로 사용할 수 있는지는 분명하지 않다. 이는 개방도가 일반적으로 여러 국가 특성과 상관관계를 가지기 때문이다. 예를 들어, 수입 수준에 따라 체계적으로 달라질 수 있다. 그림 1.1의 산점도는 이를 보여준다. 각 점은 하나의 국가를 나타내며, 곡선은 최소제곱법으로 피팅된 값이다. 곡선 아래에 위치한 국가는 해당 소득 수준에서 “정상적으로” 기대되는 것보다 무역이 적은 것으로 간주될 수 있다.

개방도가 소득 수준, 지리적 위치(예: 내륙국 여부), 규모와 같은 국가 특성과 상관관계를 가지는 것이 중요한 이유는 두 가지이다. 첫 번째는 측정과 관련이 있고, 두 번째는 논리적 이유이다.

#### 측정 문제

“순(raw)”  개방도 지표가 다른 국가 특성에 대한 정보를 포함하고 있기 때문에, 이를 조정 없이 국가 간 비교에 사용해서는 안 된다. 예를 들어, 벨기에는 미국보다 GDP 대비 무역액의 비율이 높지만, 이는 주로 미국이 더 큰 경제 규모를 가지며 스스로와의 거래가 더 많기 때문이다. 의미 있는 비교를 생성하려면 개방도 비율과 관련이 없다고 판단되는 경제 규모와 같은 영향을 통제해야 한다. 이러한 통제는 회귀 분석을 통해 수행할 수 있으며, 이에 대한 예는 아래 응용 1에서 제공될 것이다.

#### 논리적 문제

개방도가 경제성장에 미치는 영향을 계량적으로 평가하려 한다고 가정해 보자. 그러한 회귀 분석에서 개방도 지표를 설명변수로 사용할 경우, **역인과관계**(**reverse causality**; 경제성장이 개방도에 영향을 미침)나 **누락변수**(**omitted variables**; 가령 정부/제도의 질적 성격의 경우에는 개방도와 경제성장 모두에 영향을 미칠 수 있음)의 문제가 발생하지 않도록 정제되어야 한다. 그렇지 않으면, 우리가 발견하는 관계는 “<b>내생성 편향(endogeneity bias)</b>”을 겪게 될 것이다.

성장/개방도의 회귀 분석에서 내생성 편향을 제거하려면, 개방도와 상관관계를 가지지만 개방도를 통해서만 소득/성장에 영향을 미치는 “<b>도구 변수(instrumental variables)</b>”를 사용하는 <b>식별(identification)</b> 전략을 채택해야 한다. 예를 들어, Frankel과 Romer(1999)<sup>[주1](#footnote1)</sup>는 무역 상대국과의 거리 등 이른바 “중력(gravity)” 변수를 도구 변수로 사용했다. 이 전략을 통해, 그들은 개방도가 소득 수준에 긍정적인 영향을 미친다는 결과를 발견했다. 또 다른 접근법은 결과물이 아닌 정책을 기반으로 한 개방도 지표(openness based on policies)를 사용하는 것이다. 정책을 기반으로 한 개방도 지표에 대해서는 뒤에 다시 다룰 것이다.

<small> <p id="footnote1">[주1]: Frankel, J. A. and Romer, D. (1996), “[Trade and growth: an empirical investigation](https://www.nber.org/papers/w5476),” National Bureau of Economic Research Working Paper 5476, Cambridge, MA: NBER. </p> </small>

참고로, GDP 대비 무역액 비율과 같이 단순해 보이는 지표조차도 해석에 따라 의미가 다를 수 있다. 개방도가 동일하더라도, 큰 해안선을 가지고 대규모 시장과 가까운 나라와 내륙국으로 고립되어 있고 소득 수준이 낮은 나라에게는 전혀 다른 의미를 가질 수 있다.

그림 설명: 대외개방도는 GDP 대비 수출과 수입의 합으로 측정된다. 1인당 GDP는 구매력평가(PPP: Purchasing
Power Parity)에 따른 미달러화 기준이다. **그림 a**에서 곡선은 개방도를 종속변수로, 1인당 GDP를 독립변수로 하는 OLS 회귀선이다. **그림 b**에서는 1인당 GDP가 로그 변환되어 있다. 산점도의 모양이 바뀐 것을 알 수 있다. 이상값(outliers)의 영향이 감소하며, 그림 b에서도 여전히 관계는 오목한 형태를 보이지만, 전환점(turning point)은 그림 a와 동일한 1인당 GDP 수준에 있지 않다. 그림 a에서는 전환점이 PPP 20,000달러 약간 아래에 위치하며, 그림 b에서는 약 exp(9.5) = PPP 13,400달러 수준에 위치한다. 이는 정성적 결론(관계의 오목한 형태)은 강건할 수 있지만, 정량적 결론(전환점의 위치)은 추정 방법의 사소해 보이는 변경에도 크게 달라질 수 있음을 강조한다. 종합적으로, 1인당 GDP가 낮은 수준에서는 이 더 빠르게 증가하며, 높은 수준에서는 증가 속도가 완만해지는 경향이 있음을 보여준다.

1. 국가 간 개방성 비교

국가의 기초 요인에 비추어 해당 국가가 예상되는 무역량 대비 실제로 얼마나 무역을 하는지를 올바르게 측정하기 위해, 아래와 같은 무역 개방성 회귀식을 실행할 수 있다:

$$
O_i = \alpha_0 + \alpha_1 y_i + \alpha_2 LL_i + \alpha_3 R_i + u_i \tag{1.34}
$$

여기서  
- $y_i$는 1인당 GDP,  
- $LL_i$는 국가 $i$가 내륙국이면 1, 그렇지 않으면 0인 (소위 "더미 변수") 값,  
- $R_i$는 외진 정도(remoteness), 그리고  
- $u_i$는 오차항이다.

이 접근 방식은 Leamer (1988)의 연구로 거슬러 올라가며, 해당 방정식은 OLS(최소제곱법)로 추정할 수 있다.

이러한 우변 변수(RHS)를 사용함에 있어, 이미 문제가 발생한다는 점에 주목해야 한다. GDP를 현행 가격과 현행 환율로 측정해야 하는지, 아니면 구매력 평가(Purchasing-Power Parity, PPP) 수준으로 측정해야 하는지에 대한 문제가 제기된다. 이 문제에 대한 전면적인 논의는 3장에서 다루겠지만, 국가 간 GDP 측정에는 결코 사소하지 않은 문제가 내포되어 있음을 여기서 언급하는 것으로 충분하다.

$O_i$와 그 예측값 $\hat{O}_i$의 차이(잔차, residual)는 국가 $i$의 개방성을 "정리된(purged)" 척도로 해석할 수 있다. 잔차가 양수이면, 해당 국가는 그 특성에 비추어 예상보다 더 많은 무역을 하고 있는 것이며; 잔차가 음수이면, 무역량이 예상보다 적음을 의미한다.

추정 결과는 1인당 GDP가 개방성과 양의 관계에 있음을 보여준다(그림 1.1의 기술 통계 참조). 또한, 인구가 많을수록 개방성이 높고, 내륙국인 경우 개방성이 감소함을 시사한다. 설명 변수와 상관될 수 있는 누락변수(omitted variable) 문제를 줄이기 위해 국가 고정효과(country fixed effect)를 포함하는 것이 필요하며, 이는 추정에 편향을 도입할 수 있는 문제를 완화하는 데 기여한다.

- **외진 정도**: 국가나 지역이 주요 시장으로부터 물리적으로 멀리 떨어져 있는 정도를 나타내는 표현.
- **원격성**: 'remoteness'의 직역에 가까우며, 지리적 거리를 강조하는 용어.
- **지리적 고립도**: 국가가 주요 경제 중심지나 무역 상대국들로부터 얼마나 고립되어 있는지를 강조하는 표현.

## 2. 데이터

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import statsmodels.api as sm
import statsmodels.formula.api as smf
import os

# 데이터 읽기
df = pd.read_stata("../Data/openness.dta")
df
```

**변수**

- `ccode` : 국제표준화기구(ISO)의 나라 코드  
- `year` : 연도  
- `openc` : 현재가치 기준 대외개방도(%)
- `openk` : 실질가치 기준 대외개방도(%)
- `pop` : 인구(천 명)
- `gdp_current` : 현재가치 기준 GDP(미국 달러)
- `ldlock` : 내륙국 여부 더미변수(1이면 내륙국)  
- `island` : 섬 여부 더미변수(1이면 섬)  
- `remoteness` : 고립도  
- `remoteness_head` : Keith Head의 정의에 따른 고립도  

```python
# 한국 데이터만 필터링
kor_data = df[df['ccode'] == 'KOR']
kor_data
```

```python
# 연도별 'openc' 값 선그래프 그리기
plt.figure(figsize=(6,4))
plt.plot(kor_data['year'], kor_data['openc'], marker='o', linestyle='-', color='b')

# 그래프 제목 및 라벨 설정
plt.title('Openness for Korea')
plt.xlabel('Year')
plt.ylabel('Openc')

# 그래프 출력
plt.show()
```

```python
# 변수 생성 및 로그 변환
df['gdppc'] = df['gdp_current'] / df['pop']
df['ln_open'] = np.log(df['openc'])
df['ln_gdp'] = np.log(df['gdp_current'])
df['ln_gdppc'] = np.log(df['gdppc'])
df['ln_pop'] = np.log(df['pop'])
df['ln_remot'] = np.log(df['remoteness'])
df['ln_remot_head'] = np.log(df['remoteness_head'])

# GDP per capita 재조정
df['gdppc'] = df['gdppc'] / 1000
df['ln_gdppc'] = np.log(df['gdppc'])
```

```python
# 플롯: Quadratic fit
sns.set_theme(style="whitegrid")
plt.figure(figsize=(10, 6))
filtered_df = df[(df['year'] == 2000) & (df['openc'] <= 200)]
sns.scatterplot(x='gdppc', y='openc', data=filtered_df, label='Openness')
sns.regplot(x='gdppc', y='openc', data=filtered_df, order=2, scatter=False, label='Quadratic fit')

plt.title("Quadratic fit")
plt.xlabel("GDP per capita")
plt.ylabel("Openness")
plt.legend()
plt.show()
```

```python
# 플롯: Quadratic fit
sns.set_theme(style="whitegrid")
plt.figure(figsize=(10, 6))
filtered_df = df[(df['year'] == 2000) & (df['openc'] <= 200)]
sns.scatterplot(x='ln_gdppc', y='openc', data=filtered_df, label='Openness')
sns.regplot(x='ln_gdppc', y='openc', data=filtered_df, order=2, scatter=False, label='Quadratic fit')

plt.title("Quadratic fit after log transformation")
plt.xlabel("log GDP per capita")
plt.ylabel("Openness")
plt.legend()
plt.show()
```

## 3. 회귀분석

```python
# 회귀 분석 및 결과 저장
# 1차 회귀 분석
model1 = smf.ols('ln_open ~ ln_gdppc + C(ccode)', data=df).fit()
print(model1.summary())
```

**`linearmodels` 패키지 설치**

```python
#!pip install linearmodels  # 처음에 한 번만 설치
```

```python
# ccode와 year를 인덱스로 설정
data = df.set_index(['ccode', 'year'])
data = data[['ln_open', 'ln_gdppc']]
data = data.dropna()
data
```

```python
from linearmodels.panel import PooledOLS, PanelOLS
from statsmodels.api import add_constant

# 고정효과 모델을 사용하여 회귀 모델을 실행
mod = PanelOLS(data['ln_open'], data['ln_gdppc'], 
               drop_absorbed=True, entity_effects=True)
fe_results = mod.fit(cov_type='clustered', cluster_entity=True)

# 결과 요약을 출력
print(fe_results)
```

```python
# 회귀 분석 및 결과 저장
# 1차 회귀 분석
model1 = smf.ols('ln_open ~ ln_gdppc', data=df).fit()
print(model1.summary())
```

```python
from linearmodels.panel import PooledOLS, PanelOLS
from statsmodels.api import add_constant

# X 변수
X_with_const = add_constant(data['ln_gdppc'])

# Pooled OLS 모델을 사용하여 회귀 모델을 실행
mod = PooledOLS(data['ln_open'], X_with_const)
pooled_results = mod.fit(cov_type='clustered', cluster_entity=True)

# 결과 요약을 출력
print(pooled_results)
```

```python
# 2차 회귀 분석 (ln_pop 추가)
model2 = smf.ols('ln_open ~ ln_gdppc + ln_pop + C(ccode)', data=df).fit()
print(model2.summary())
```

```python
# 3차 회귀 분석 (ldlock 추가)
model3 = smf.ols('ln_open ~ ln_gdppc + ln_pop + ldlock + C(ccode)', data=df).fit()
print(model3.summary())
```

```python
# 4차 회귀 분석 (ln_remot_head 추가)
model4 = smf.ols('ln_open ~ ln_gdppc + ln_pop + ldlock + ln_remot_head + C(ccode)', data=df).fit()
print(model4.summary())
```

```python
# 결과를 요약하여 표로 작성
columns = ["(1)", "(2)", "(3)", "(4)"]
rows = [
    "ln_gdppc",
    "ln_pop",
    "ldlock",
    "ln_remot_head",
    "Constant",
    "Observations",
    "R-squared"
]

# 결과값 저장 함수
def extract_results(models):
    data = []
    for model in models:
        coef = model.params
        std_err = model.bse
        obs = int(model.nobs)
        r2 = model.rsquared

        # Coefficient + SE for each variable
        row_data = {
            "ln_gdppc": f"{coef.get('ln_gdppc', 'N/A'):.4f} ({std_err.get('ln_gdppc', 'N/A'):.4f})" if 'ln_gdppc' in coef else "",
            "ln_pop": f"{coef.get('ln_pop', 'N/A'):.4f} ({std_err.get('ln_pop', 'N/A'):.4f})" if 'ln_pop' in coef else "",
            "ldlock": f"{coef.get('ldlock', 'N/A'):.4f} ({std_err.get('ldlock', 'N/A'):.4f})" if 'ldlock' in coef else "",
            "ln_remot_head": f"{coef.get('ln_remot_head', 'N/A'):.4f} ({std_err.get('ln_remot_head', 'N/A'):.4f})" if 'ln_remot_head' in coef else "",
            "Constant": f"{coef.get('Intercept', 'N/A'):.4f} ({std_err.get('Intercept', 'N/A'):.4f})",
            "Observations": obs,
            "R-squared": f"{r2:.3f}",
        }
        data.append(row_data)
    return data

# 모델 결과 추출
model_results = extract_results([model1, model2, model3, model4])

# 데이터프레임 생성
result_df = pd.DataFrame(model_results, index=columns)

# 표를 원하는 형식으로 변환
table = pd.DataFrame(index=rows)
for col, model in zip(columns, model_results):
    table[col] = [
        model.get("ln_gdppc", ""),
        model.get("ln_pop", ""),
        model.get("ldlock", ""),
        model.get("ln_remot_head", ""),
        model.get("Constant", ""),
        model.get("Observations", ""),
        model.get("R-squared", ""),
    ]

# 표 출력
table

# 표를 LaTeX, Excel 또는 CSV로 저장하려면 아래 코드 사용 가능
# table.to_csv("regression_results.csv")
# table.to_latex("regression_results.tex")
```

```python
# Dummy 생성: 예측값 대비 관측값 비교
df['fitted'] = model1.predict(df)
df['trade_more'] = (df['ln_open'] > df['fitted']).astype(int)

# 국가별 trade_more 집계
trade_more_summary = df.groupby('ccode')['trade_more'].value_counts().unstack(fill_value=0)
print(trade_more_summary)
```
