## 1. 비관세조치

비관세조치(Non-Tariff Measures, NTM)는 관세 외에 무역에 영향을 미치는 모든 정부 정책을 뜻한다. 즉, 상품을 수입하거나 수출할 때 세금(관세) 외에 기업이 따라야 하는 규정, 요건, 제한 등을 포함하는 모든 조치이다.  

**왜 비관세조치가 중요한가?**  

비관세조치는 단순히 무역을 제한하는 것이 아니라, 소비자 보호, 환경 보호, 공정 무역 유지 등의 이유로 필요하다. 예를 들어:  
- 안전 규정: 전자제품이 화재 위험 없이 안전하게 사용될 수 있도록 요구하는 기술 기준  
- 보건 및 위생 규정: 수입 식품이 식중독을 유발하지 않도록 특정 위생 기준을 충족해야 하는 경우  
- 환경 보호: 플라스틱 제품에 대한 규제나 오염 물질 배출 기준  

하지만 비관세조치가 과도하면 기업들이 해외 시장에 진출하기 어려워지고, 소비자들은 더 높은 가격을 지불해야 할 수도 있다.  

**비관세조치의 종류**

비관세조치는 매우 다양하지만, 대표적인 유형은 다음과 같다:  

- 기술적 조치 (Technical Measures)  
   - 제품 안전 기준, 환경 보호 규정, 위생·검역 조치(SPS) 등  
   - 예: 특정 농산물의 농약 잔류 기준을 충족해야만 수출 가능  

- 수량 제한 (Quantity Restrictions)  
   - 특정 제품의 수입·수출량을 제한하는 조치  
   - 예: 자동차 수입 쿼터(일정 수 이상 수입 불가)  

- 원산지 규정 (Rules of Origin)  
   - 제품이 특정 국가에서 생산되었음을 증명해야 하는 규칙  
   - 예: "Made in Korea"라고 표시하려면 국내에서 일정 비율 이상 제조해야 함  

- 세이프가드 및 반덤핑 조치  
   - 특정 국가에서 너무 저렴한 가격에 수입되는 제품(덤핑)에 대한 추가 관세 부과  
   - 예: 철강 제품이 특정 국가에서 너무 싸게 들어오면 국내 산업 보호를 위해 추가 세금 부과  

**결론**

비관세조치는 무역에 중요한 역할을 한다.  
- 긍정적인 측면: 소비자 보호, 환경 보호, 공정 경쟁 유지  
- 부정적인 측면: 과도한 규제는 기업 부담 증가, 시장 접근 어려움  

따라서 각국 정부와 국제기구(WTO, UNCTAD)는 비관세조치의 투명성을 높이고 공정한 규정을 만들기 위해 노력하고 있다.

## 2. 연구자를 위한 NTM 데이터  

출처: [Non-Tariff Measures Data for Researchers](https://trainsonline.unctad.org/bulkDataDownload)

비관세조치(NTM)는 관세 외에 무역에 영향을 미칠 수 있는 정책으로, 보건·안전·환경 보호뿐만 아니라 무역 규제에도 중요한 역할을 합니다. 이러한 조치에는 독성 물질의 최대 허용 기준과 같은 안전 소비를 위한 제품 요구사항, 수입 할당량 및 원산지 규칙과 같은 전통적인 무역 정책이 포함됩니다.  

[UNCTAD 비관세조치 프로그램](https://unctad.org/topic/trade-analysis/non-tariff-measures)에서는 [NTM에 해당하는 정책의 범위](https://unctad.org/topic/trade-analysis/non-tariff-measures/NTMs-classification)를 정의하고, [공통된 NTM 분류 체계를 개발](https://unctad.org/webflyer/international-classification-non-tariff-measures-2019-version)하며, 데이터를 [수집](https://unctad.org/topic/trade-analysis/non-tariff-measures/NTMs-data-collection)·[공유](https://unctad.org/topic/trade-analysis/non-tariff-measures/NTMs-data-dissemination)하고, 투명성과 우수한 규제 관행 강화를 위한 [정책 지원](https://unctad.org/topic/trade-analysis/non-tariff-measures/NTMs-policy-support)을 제공합니다. NTM에 대한 개요는 [비관세조치 A to Z](https://unctad.org/webflyer/non-tariff-measures-z)에서도 확인할 수 있습니다.  

[NTM 연구](https://unctad.org/topic/trade-analysis/non-tariff-measures/NTMs-analysis)는 무역과 [지속가능발전목표](https://unctad.org/topic/trade-analysis/non-tariff-measures/NTMs-and-SDGs)(SDG: Sustainable Development Goals)에서 NTM의 높은 중요성을 보여줍니다. 연구자들은 [UNCTAD TRAINS Database](https://unctad.org/topic/trade-analysis/trains)를 무료로 이용할 수 있으며, 연구 출판물이 있을 경우 UNCTAD에 통지할 것을 권장합니다. 데이터 출처는 반드시 명확히 명시해야 하며(예: *UNCTAD TRAINS Database*), [이용 약관](https://unctad.org/terms-use)을 준수해야 합니다.  

UNCTAD TRAINS 데이터베이스에는 NTM 및 [관세 데이터](https://unctad.org/topic/trade-analysis/data-statistics-and-trends)가 포함되어 있습니다. 이 [TRAINS 포털](https://trainsonline.unctad.org/#)에서는 제품(HS 분류), 경제권, 지역, NTM 유형별로 쉽게 접근할 수 있는 정보를 제공합니다. 120개 이상의 경제권을 대상으로 하며, 여러 연도의 데이터를 포함하는 경우가 많습니다.  

데이터 수집 과정은 [NTM 데이터 수집 지침](https://unctad.org/webflyer/guidelines-collection-data-official-non-tariff-measures)에서 확인할 수 있으며, 데이터 구조에 대한 설명은 [UNCTAD TRAINS: 글로벌 비관세조치 데이터베이스](https://unctad.org/webflyer/unctad-trains-global-database-non-tariff-measures)에서 확인할 수 있습니다.  

#### UNCTAD TRAINS

UNCTAD의 TRAINS(Trade Analysis and Information System)은 무역, 관세, 비관세조치(NTM) 데이터를 통합한 무역 및 시장 접근 정보 시스템입니다. TRAINS에는 170개 이상의 국가에 대한 HS 기반 관세 데이터가 여러 연도에 걸쳐 포함되어 있습니다.  

TRAINS는 데이터 저장소이자 분석 도구로, 국제 상품 무역에 참여하는 정책 입안자 및 경제 운영자들을 위해 설계되었습니다. 또한 다자 및 양자 무역 협상을 위한 강력한 도구이며, 국제 상품 무역에 대한 일반적인 연구에도 활용됩니다.  

TRAINS의 **관세 및 무역 데이터**는 UNCTAD와 세계은행이 공동 개발 및 유지하는 **[WITS ](https://wits.worldbank.org/)**(World Integrated Trade Solution) 애플리케이션을 통해 제공됩니다. TRAINS의 **비관세조치(NTM) 데이터** 역시 다양한 플랫폼을 통해 제공됩니다.

#### TRAINS 포털

[**TRAINS Portal**](https://trains.unctad.org/)은 전 세계의 무역 관련 규정 및 비관세조치(NTM) 데이터를 제공하는 대화형 웹 포털로, 공공 사용자들이 효율적인 검색, 데이터 수집, 데이터 검색뿐만 아니라 맞춤형 검색 결과 다운로드 기능을 활용할 수 있도록 지원합니다.  

이 웹사이트에서 제공되는 NTM 데이터는 **TRAINS Data Entry Tool**를 통해 수집된 것으로, 109개 이상의 경제권에서 표준화된 접근 방식을 사용하여 가장 포괄적인 무역 규정 및 NTM 데이터를 확보하였으며, 경제권 간 비교 가능성을 보장합니다.

### 1. 서론  

TRAINS NTM 데이터베이스는 100개 이상의 국가에 대한 포괄적인 데이터를 포함하고 있다. 이 데이터는 특정 시점에서 특정 국가, 특정 제품, 특정 교역 상대국에 영향을 미칠 수 있는 모든 요구사항을 다루며, 이를 통해 무역 관련 정보를 투명하게 제공하고 통계 분석을 용이하게 만들어 경제 연구에 유용한 자료가 된다.  

TRAINS NTM 데이터베이스는 제품, 조치 유형, 조치를 부과하는 국가, 영향을 받는 국가 등 다양한 변수로 분류된 체계적인 정보를 제공한다. 따라서 이 데이터는 특히 중소기업(SME) 무역업자들의 수출을 촉진하고, 무역 정책을 설계하는 정부 관계자들에게도 가치 있는 도구가 된다.  

이 데이터는 일괄(bulk) 다운로드가 가능하며, 추가적인 가공 및 클리닝 과정을 거쳐 통계 분석에 보다 일관성 있고 적합한 형태로 제공된다. 일괄 다운로드를 통해 제공되는 NTM 데이터는 발생 빈도와 같은 기술적 통계(descriptive statistics)를 생성하는 데 활용될 수 있으며, 이는 보다 복잡한 평가 모델에서 변수로도 사용될 수 있다.  

본 문서는 원시 데이터의 클리닝 과정과 이를 통해 생성된 변수들을 설명한다. 특히, UNCTAD TRAINS 웹사이트에서 일괄 다운로드할 수 있는 연구자 파일(Researcher file)에 초점을 맞추고, 해당 데이터의 클리닝 절차를 상세히 기술한다. 이 출판물은 UNCTAD의 기존 NTM 관련 연구를 기반으로 한다.  

원본 데이터 수집 과정, 데이터 설명, 이용 가능한 정보에 대한 보다 자세한 내용은 **TRAINS 사용자 가이드(UNCTAD, 2017)**에서 확인할 수 있다. 동일한 NTM 데이터는 다음과 같은 다양한 형식으로 제공된다.  

- WITS (World Integrated Trade Solution)  
- 웹 애플리케이션 TRAINS(Trade Analysis Information System)  
- Researcher file for bulk download  
- Global Trade Helpdesk  

클리닝된 데이터셋을 기반으로 통계 지표(발생 빈도 등)가 계산되며, 이는 이전 UNCTAD 보고서에서 제시한 방식에 따라 공개된다.  

데이터 클리닝 과정을 논의할 때, 이는 원본 TRAINS 데이터가 품질 관리가 부족하거나 정리가 안 된다는 의미가 아님을 명확히 할 필요가 있다. 다만, 특정 세부사항(예: 세율 수준의 정보 또는 중복 조치 문제)이 국가 간 비교를 어렵게 만들 수 있음을 고려한 것이다. 또한, 데이터베이스에 포함된 NTM이 각국의 법적 요구사항에 따른 것이라 하더라도, 이는 해당 요구사항의 정당성이나 적절성에 대한 판단을 의미하지 않는다. NTM은 국제 무역에 영향을 미칠 수 있는 정책 도구의 투명성을 높이기 위한 목적으로 중립적으로 기록된다.

### 2. 비관세장벽(NTM) 데이터  

본 문서에서 제공하는 NTM 데이터는 앞서 언급한 바와 같이 다른 온라인 도구를 통해서도 제공된다. UNCTAD TRAINS 웹사이트에서는 이용 가능한 NTM 데이터를 검색할 수 있으며, 같은 사이트에서 STATA 및 CSV 형식의 일괄 다운로드 파일도 제공된다. 이 파일은 각 국가에서 이용 가능한 모든 연도의 HS 6단위 수준 데이터를 포함하며, 본 문서에서 설명하는 절차에 따라 정제된 데이터를 제공한다.  

NTM의 분류 체계, 데이터 수집 방법, 데이터의 강점과 한계에 대한 보다 자세한 내용은 UNCTAD (2017)에 기술되어 있다. 해당 문서에서 설명하는 변수들은 본 데이터베이스에 포함된 변수들과 일치하며, HS 6단위 수준으로 집계되어 있지만, 본 문서에서 설명하는 정제 과정을 거치기 전의 원본 데이터를 포함하고 있다.  

통계적 활용을 위해 제공되는 데이터는 CSV 및 STATA 형식으로 제공되며, 분석 단위는 다음 다섯 가지 요소의 조합을 기반으로 한다.  

1. 비관세장벽(NTM)을 시행하는 국가(Reporter)  
2. 교역 상대국(Partner)  
3. HS 6단위 수준에서 영향을 받는 제품  
4. 최대 4자리까지 세분화된 NTM 코드  
5. 데이터 수집 연도  

이 다섯 가지 조합은 데이터베이스에서 한 번만 기록된다. 변수 `nbr`은 원본 데이터에서 동일한 조합이 몇 번 발견되었는지를 나타낸다. 이 변수는 동일한 NTM 코드가 두 개 이상의 규제에서 동일한 제품에 적용된 경우를 반영한다.  

본 문서는 패널 데이터를 중심으로 논의하며, 모든 국가에서 매년 데이터를 수집한 것은 아님을 인지해야 한다. 일부 국가는 매년 데이터가 존재하지만, 일부 국가는 특정 시점에 한두 개의 데이터만 존재할 수도 있다. 또한, 규제의 발행 연도와 데이터 수집 연도를 혼동해서는 안 된다. 예를 들어, 특정 국가 X의 데이터가 2020년에 수집되었다면, 데이터베이스에는 해당 연도에 시행된 모든 규제가 포함된다. 하지만 이 규제 중 일부는 2010년부터 시행된 것일 수도 있으며, 여전히 무역에 영향을 미친다면 2020년 데이터베이스에 포함된다. 그러나 이는 2010년의 NTM 데이터를 포함하고 있다는 의미는 아니다.

### 3. 원시 데이터 클리닝  

NTM 데이터를 클리닝하는 과정은 경제 분석에 유용한 NTM 지표를 도출하기 위한 작업이다. ‘클린’ NTM 데이터를 사용하면, NTM의 발생 빈도나 보급률을 통해 특정 국가의 규제 관행을 평가하는 데 효과적이다. 반면, 특정 제품이나 정책 도구에 대한 세부적인 정보를 원하는 분석가들은 원본 TRAINS NTM 데이터를 참고해야 한다. 특히, 규제가 요구하는 세부 사항을 깊이 분석하려는 경우 원시 데이터가 더 적합하다.  

STATA 파일은 일반적인 통계 분석을 위해 설계되었으며, 이러한 목적에 맞게 구성되어 있다. 따라서 데이터 클리닝은 첫 번째 필수 단계가 된다.  

#### 수평적 조치(Horizontal Measures) 및 부분 적용(Partial Coverage)  

어떤 조치가 모든 HS 코드에 동시에 영향을 미친다면 이를 수평적 조치(horizontal measure)라고 한다. 예를 들어, 기업이 수입 또는 수출을 허가받기 위해 반드시 무역업자로 등록해야 하는 경우, 이는 제품의 종류와 관계없이 모든 무역 활동에 적용되므로 수평적 조치로 간주된다.   

부분 적용(Partial Coverage)이란 특정 조치가 제품 전체에 적용되지 않고, 특정 조건에서만 적용되거나 가장 세분화된 제품 코드의 일부 하위 범주에만 영향을 미치는 경우를 의미한다. 이러한 경우, 해당 HS6 제품 코드는 '부분 적용'으로 표시되며, ‘완전 적용(fully affected)’되는 경우와 구별된다. 부분 적용은 HS6 코드 내에서도 특정 관세품목 코드(8자리 이상)에는 적용되고, 다른 코드에는 적용되지 않을 수 있다. 이에 대한 보다 자세한 내용은 UNCTAD의 "Guidelines for the Collection of Data on Official Non-Tariff Measures"을 참고할 수 있다.  

데이터 수집 과정에서 부분 적용과 완전 적용은 모두 플래그 처리되며, 본 파일에도 그대로 반영된다. 제공된 데이터 파일은 수평적 조치 및 부분 적용된 조치를 포함하고 있으며, 이 여부가 명확하게 표시되어 있다. 기존 UNCTAD에서 발표한 NTM 발생 빈도 지표는 본 데이터의 일부만 사용하였으며, 완전 적용된 NTM만을 고려하고, 수평적 조치는 제외하였다. 반면, 현재 제공되는 파일은 보다 포괄적인 데이터를 포함하여, 분석가가 필요에 따라 특정 범주의 데이터를 제외할 수 있도록 구성되어 있다.  

통계 분석에서 수평적 조치를 제외하는 것은 필수적이다. 만약 수평적 조치를 포함하면, 빈도 지수(Frequency Index)와 적용 비율(Coverage Ratio)이 1이 되어버려 국가의 규제 패턴을 왜곡할 수 있다. 즉, 일부 행정적 성격을 가진 수평적 조치 몇 개만으로도 모든 제품이 규제를 받는 것처럼 보일 수 있으며, 이는 NTM의 다양성을 반영하지 못하게 된다. 따라서 수평적 조치를 제외함으로써 보다 정교한 규제 환경 분석이 가능해진다.

### 4. 최종 '클린’ 데이터셋  

다운로드 가능한 데이터셋은 115개 보고국(유럽연합 소속 국가들은 하나의 보고국으로 간주)을 포함한다. 최신 연도 데이터만을 포함하는 횡단면(cross-section) 하위 데이터셋에는 보고국, HS6 코드, NTM 코드, 교역 상대국의 조합으로 이루어진 7,700만 개 이상의 관측값이 포함되어 있다. 그러나 전체 데이터셋은 단일 연도뿐만 아니라, 다년간의 데이터를 보유한 국가들의 모든 가용 연도를 포함하고 있다.  

#### 데이터 구조 및 주요 변수  

각 행(row)은 보고국, 교역 상대국, HS6 제품 코드, NTM 코드, 데이터 수집 연도의 조합을 나타낸다.  

추가적으로, 특정 제품에 대해 여러 개의 인증 요건이 적용될 수 있으므로, 원본 데이터에서 동일한 조합이 몇 번 발생했는지를 나타내는 변수도 포함되어 있다. 예를 들어, 동일한 제품에 대해 두 개의 다른 인증 요건이 있을 수 있으므로, 해당 제품에 적용되는 NTMs의 개수가 1보다 클 수도 있다.  

Researcher 파일에서는 NTM의 특성을 고려하여 원본 데이터에서의 발생 빈도를 계산한 여러 변수를 제공한다. 주요 변수는 다음과 같다.  

- ntm_all: 모든 유형의 NTM을 포함한 총 개수  
- ntm_FullCoverage: 특정 제품에 완전히 적용되는 NTM 개수  
- ntm_PartialCoverage: 특정 제품에 부분적으로 적용되는 NTM 개수  
  - 두 변수 모두 수평적 조치(horizontal measures)와 비수평적 조치(non-horizontal measures)를 구분하지 않고 합산함  

또한, NTM이 수평적 조치(horizontal measure)인지 아닌지를 기준으로 구분한 변수도 있다.  
- ntm_H: 수평적 조치의 개수  
- ntm_nonH: 비수평적 조치의 개수  
  - 이 두 변수는 NTM이 부분 적용되는지 완전 적용되는지에 대한 구분 없이 전체 개수를 제공함  

보다 세부적인 구분을 위해, 아래의 추가 변수가 제공된다.  
- ntm_FullCov_H: 수평적 조치 중 모든 제품에 완전히 적용되는 NTM 개수  
- ntm_PartialC_H: 수평적 조치 중 모든 제품에 부분적으로 적용되는 NTM 개수  
- ntm_FullCov_nonH: 수평적 조치를 제외한, 특정 제품에 완전히 적용되는 NTM 개수  
  - 과거 UNCTAD 분석에서 주요 변수로 사용됨  
- ntm_PartialC_nonH: 수평적 조치를 제외한, 특정 제품에 부분적으로 적용되는 NTM 개수

데이터의 크기가 너무 커서 메모리 할당에 실패

**배치 처리**: 데이터셋이 매우 클 경우, 데이터를 한 번에 처리하지 않고 배치 단위로 처리할 수 있습니다. `pandas`의 `chunksize` 매개변수를 사용하면 파일을 작은 덩어리로 나누어 처리할 수 있습니다.

**필터링 후에 데이터프레임 메모리 해제**: 필터링을 한 후 `df`가 여전히 메모리에 남아 있을 수 있으므로, `df`를 삭제하고 가비지 컬렉션을 강제로 호출하여 메모리 문제를 해결할 수 있습니다.

```python
import pandas as pd  
import time

file_path = r"C:\Comtrade\Data\NTM_hs6_2010_2022_H4V12.csv"  

start_time = time.time()

chunk_size = 1000000  # 배치 크기 설정 (1백만 행 단위)
df_filtered_list = []  # 필터링된 결과를 저장할 리스트

# 파일을 배치 단위로 읽어오기
chunks = pd.read_csv(file_path, chunksize=chunk_size)

for chunk in chunks:
    # 각 배치에서 'ReporterISON'이 410인 행만 필터링
    df_filtered_chunk = chunk[chunk['ReporterISON'] == 410]
    
    # 필터링된 데이터를 리스트에 추가
    df_filtered_list.append(df_filtered_chunk)

# 모든 필터링된 데이터를 하나로 합치기
df_kor = pd.concat(df_filtered_list, ignore_index=True)

end_time = time.time()

elapsed_time = end_time - start_time
minutes = int(elapsed_time // 60)
seconds = elapsed_time % 60

print(f"전체 데이터 로딩 시간: {minutes}분 {seconds:.2f}초")
```

```python
import pandas as pd  
import time

file_path = r"C:\Comtrade\Data\NTM_hs6_2010_2022_H4V12.csv"  

start_time = time.time()

chunk_size = 1000000  # 배치 크기 설정 (1백만 행 단위)
df_filtered_list = []  # 필터링된 결과를 저장할 리스트

# 파일을 배치 단위로 읽어오기
chunks = pd.read_csv(file_path, chunksize=chunk_size)

for chunk in chunks:
    # 각 배치에서 'ReporterISON'이 410인 행만 필터링
    df_filtered_chunk = chunk[chunk['ReporterISON'] == 410]
    
    # 필터링된 데이터를 리스트에 추가
    df_filtered_list.append(df_filtered_chunk)

# 모든 필터링된 데이터를 하나로 합치기
df_kor = pd.concat(df_filtered_list, ignore_index=True)

end_time = time.time()

elapsed_time = end_time - start_time
minutes = int(elapsed_time // 60)
seconds = elapsed_time % 60

print(f"전체 데이터 로딩 시간: {minutes}분 {seconds:.2f}초")

# 결과 저장 및 확인
df_kor.to_csv('../Data/NTM_KOR.csv', index=False)
df_kor
```
